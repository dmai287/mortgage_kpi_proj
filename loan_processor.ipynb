```python
import pandas as pd
import numpy as np
```

```python
loan_df = pd.read_excel('Encompass Data Feb 16 2023.xlsx')
loan_df = loan_df.loc[loan_df['Active Loan']=='No']
loan_df.shape
```

```python
loan_df['Month'] = loan_df['Date Completed'].astype(str)
loan_df['Month'] = loan_df['Month'].apply(lambda x: x[5:7])
loan_df['Year'] = loan_df['Date Completed'].astype(str)
loan_df['Year'] = loan_df['Year'].apply(lambda x: x[0:4])
```

```python
#Fill the empty Month by Loan Created Date value
loan_df['Loan Created Date'] = pd.to_datetime(loan_df['Loan Created Date'])

loan_df['Month'] = np.where((loan_df['Month'] == '') | (loan_df['Month'].isna()),loan_df['Loan Created Date'].dt.month+1,loan_df['Month'])
loan_df['Year'] = np.where((loan_df['Year'] == ' ') | (loan_df['Year'].isna()),loan_df['Loan Created Date'].dt.year, loan_df['Year'])
```

```python
#cast format to string
loan_df['Month'] = loan_df['Month'].astype(str)
loan_df['Year'] = loan_df['Year'].astype(str)
```

```python
month_map = {
    '1': '1', '01': '1',
    '2': '2', '02': '2',
    '3': '3', '03': '3',
    '4': '4', '04': '4',
    '5': '5', '05': '5',
    '6': '6', '06': '6',
    '7': '7', '07': '7',
    '8': '8', '08': '8',
    '9': '9', '09': '9',
    '10': '10',
    '11': '11',
    '12': '12'
}
loan_df['Month'] = loan_df['Month'].map(month_map)
```

```python
loan_df
```

```python
cols_to_show = ['Year','Month','Total Loan Amount','Loan Processor']
loan_df = loan_df[cols_to_show]
loan_df = loan_df.loc[loan_df['Total Loan Amount'] != 0]
loan_df
```

```python
loan_df = loan_df.sort_values('Month').reset_index(drop=True)
loan_df.head()

# loan_df.to_csv('loan_status_full')
```
