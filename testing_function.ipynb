```python
df_fullyear_month = df_fullyear.groupby('Month Text').mean().reset_index()
df_fullyear_month
```

```python
cols_to_show = ['Month Text','Work Completion Score','Final Score', 'Work Score (%)', 'Work Completion Score (%)',
                 'Attitude Score (%)', 'Final Score (%)','Month']

df_fullyear_month = df_fullyear_month[cols_to_show]
df_fullyear_month = df_fullyear_month.sort_values(by='Month').reset_index().drop(columns='index')
df_fullyear_month
```

```python
df_loan_status = pd.read_csv('Loan Status.csv')
df_loan_status_train = df_loan_status[['Month','Loan Amount']].iloc[0:12]
df_loan_status_test = df_loan_status[['Month','Loan Amount']].iloc[12:17]
df_loan_status_test
```

```python
df_fullyear_sale_goal_train = pd.merge(df_fullyear_month,df_loan_status_train,left_on='Month Text',right_on='Month',how='inner')
df_fullyear_sale_goal_test = pd.merge(df_fullyear_month,df_loan_status_test,left_on='Month Text',right_on='Month',how='inner')
df_fullyear_sale_goal_train
```

```python
df_fullyear_sale_goal_test
```

```python
X_train = df_fullyear_sale_goal_train.drop(columns=['Month_x','Month_y','Loan Amount'])
y_train = df_fullyear_sale_goal_train['Loan Amount']
X_test = df_fullyear_sale_goal_test.drop(columns=['Month_x','Month_y','Loan Amount'])
y_test = df_fullyear_sale_goal_test['Loan Amount']
```

```python
from lazypredict.Supervised import LazyRegressor
reg = LazyRegressor(verbose=1, ignore_warnings=False,predictions=True)
models, predictions = reg.fit(X_train,X_test, y_train, y_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures,OneHotEncoder,StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split

preprocessor = ColumnTransformer(
    transformers=[
        ('ohe', OneHotEncoder(), ['Month Text']),
        ('std_scale', StandardScaler(),X_train.select_dtypes('number').columns)
    ])

X_train_pro = preprocessor.fit_transform(X_train)
X_test_pro = preprocessor.transform(X_test)
```

```python
param = {
        'poly__degree' : [1,2,3]
        }

pipeline = Pipeline(steps=[        
            ('poly', PolynomialFeatures()),
            ('lr', LinearRegression())
])

gridsearch = GridSearchCV(pipeline, param, cv=3, verbose=1)
gridsearch.fit(X_train_pro, y_train_pro)
```

```python
gridsearch.best_params_
```

```python
lr = LinearRegression()
lr.fit(X_train_pro,y_train_pro)

lr.coef_
```

```python
loan_df = pd.read_excel('Encompass Data Feb 16 2023.xlsx')
loan_df = loan_df.loc[loan_df['Active Loan']=='No']
loan_df.shape
```

```python
loan_df['Month'] = loan_df['Date Completed'].astype(str)
loan_df['Month'] = loan_df['Month'].apply(lambda x: x[5:7])
```

```python
loan_df.head()
```

```python
#Fill the empty Month by Loan Created Date value
loan_df['Month'] = loan_df['Month'].replace('',np.nan).fillna(loan_df['Loan Created Date'].apply(lambda x: x.split('/')[0]))
```

```python
loan_df['Total Loan Amount'].isna().sum()
```

```python
cols_to_show = ['Month','Total Loan Amount','Loan Processor']
loan_df = loan_df[cols_to_show]
loan_df = loan_df.loc[loan_df['Total Loan Amount'] != 0]
```

```python
month_map = {
    '1': 1, '01': 1,
    '2': 2, '02': 2,
    '3': 3, '03': 3,
    '4': 4, '04': 4,
    '5': 5, '05': 5,
    '6': 6, '06': 6,
    '7': 7, '07': 7,
    '8': 8, '08': 8,
    '9': 9, '09': 9,
    '10': 10,
    '11': 11,
    '12': 12
}
loan_df['Month'] = loan_df['Month'].map(month_map)
```
